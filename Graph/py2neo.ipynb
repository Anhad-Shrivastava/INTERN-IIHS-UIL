```python
import pandas as pd
from py2neo import Node, Relationship,Graph
```

```python
df=pd.read_csv("test1.csv")
```

'test.csv' contains sample data to create a graph database using Neo4j

```python
df.info()
df.head(10)
```

#### create a graph using neo4j desktop or enterprise edition .Use the password that you created to establish a connection with the graph 

```python
graph = Graph(password="your-password")
```

```python
for index, row in df.iterrows():
    house=df.iloc[index,0]
    houseno=str(df.iloc[index,1])
    industryname=df.iloc[index,2]
    road=df.iloc[index,3]
    suburb=df.iloc[index,4]
    city=df.iloc[index,5]
    citydistrict=df.iloc[index,6]
    state=df.iloc[index,7]
    pincode=str(df.iloc[index,8])
    lat=str(df.iloc[index,9])
    lon=str(df.iloc[index,10])
    nic=str(df.iloc[index,11])
    graph.run("MERGE (:house{house:\""+house+"\",houseno:\""+houseno+"\",industryname:\""+industryname+"\",lat:\""+lat+"\",lon:\""+lon+"\",nic:\""+nic+"\"})")
    graph.run("MERGE (:road{road:\""+road+"\",lat:\""+lat+"\",lon:\""+lon+"\"})")
    graph.run("MERGE (:suburb{suburb:\""+suburb+"\"})")
    graph.run("MERGE (:city{city:\""+city+"\",citydistrict:\""+citydistrict+"\",state:\""+state+"\"})")
    graph.run("MATCH (h:house),(r:road) WHERE h.house=\""+house+"\" and r.road=\""+road+"\" MERGE (h)-[:locatedin]->(r)")
    graph.run("MATCH (s:suburb),(r:road) WHERE s.suburb=\""+suburb+"\" and r.road=\""+road+"\" MERGE (s)<-[:locatedin]-(r)")
    graph.run("MATCH (s:suburb),(c:city) WHERE s.suburb=\""+suburb+"\" and c.city=\""+city+"\" MERGE (s)-[:locatedin]->(c)")
```
